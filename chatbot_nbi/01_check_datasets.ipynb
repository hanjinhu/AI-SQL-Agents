Download NBI ASCII files 2025: https://www.fhwa.dot.gov/bridge/nbi/ascii2025.cfm

This notebook focuses on the dataset: NBI ASCII files for Alaska.

```python
import pandas as pd
import duckdb as db
```

Next, we will load the dataset from a CSV file from the `data` folder:

```python
nbi_2025_ak = pd.read_csv("../data/nbi/AK25.csv")
nbi_2025_ak['COUNTY_CODE_003'] = nbi_2025_ak['COUNTY_CODE_003'].astype(str)
```

## Querying the Data with SQL

We will use the DuckDB library to query the data with native SQL queries using the `sql` function. For example:

```python
db.sql("DESCRIBE TABLE nbi_2025_ak")
```

Another example, the following query calculate the top 10 states by total sales:

```python
query = """
SELECT
  COUNTY_CODE_003, COUNT(*) AS BridgeCount
FROM 
  nbi_2025_ak
GROUP BY 
  COUNTY_CODE_003
ORDER BY
  BridgeCount DESC
LIMIT 10;
"""
```

```python
db.sql(query)
```
